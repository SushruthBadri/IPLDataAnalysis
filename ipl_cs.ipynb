```python
#replacing missing values with 0
count = 0
for ele in dfbbb['Batsman_Scored']:
    if(ele not in [0,1,2,3,4,5,6]):
        dfbbb['Batsman_Scored'][count]=0
    count = count + 1
        
    
```

```python
#Top 10 Batsman: 
#based on Runs scored
print('top 10 batsmen based on Runs')
topScoreList=list(dfbbb.groupby('Striker_Id').agg({'Batsman_Scored':'sum'}).sort_values(by='Batsman_Scored', ascending = False).head(10).index)
for pid in topScoreList:
    print(list(dfplayer[dfplayer['Player_Id']==pid]['Player_Name']))
```

```python
#top 10 Batsman based on 6s
print('top 10 batsmen based on 6s')
topScoreList=list(dfbbb[dfbbb['Batsman_Scored']==6].groupby('Striker_Id').agg({'Batsman_Scored':'sum'}).sort_values(by='Batsman_Scored', ascending = False).head(10).index)
for pid in topScoreList:
    print(list(dfplayer[dfplayer['Player_Id']==pid]['Player_Name']))
```

```python
#top 10 Batsman based on Number of 4s
print('top 10 batsmen based on 4s')
topScoreList=list(dfbbb[dfbbb['Batsman_Scored']==4].groupby('Striker_Id').agg({'Batsman_Scored':'sum'}).sort_values(by='Batsman_Scored', ascending = False).head(10).index)
for pid in topScoreList:
    print(list(dfplayer[dfplayer['Player_Id']==pid]['Player_Name']))
```

```python
#top 10 Batsman based on Number of 100s
dfplayerscores = dfbbb.groupby(['Match_Id','Striker_Id']).agg({'Batsman_Scored':'sum'})
```

```python
pids = []
for tup in list(dfplayerscores[dfplayerscores['Batsman_Scored']>99].index):
    pids.append(tup[1])
centurycount=dict([(pid,pids.count(pid)) for pid in pids])
```

```python
centurycount
```

```python
cent_keys = list(centurycount.keys())
sorted_vals = list(centurycount.values())
sorted_vals.sort(reverse=True)
sorted_cent = {}
for val in sorted_vals:
    for key in cent_keys:
        if centurycount[key]==val:
            sorted_cent[key]=val
```

```python
print('top 10 batsmen based on centuries')
for pid in list(sorted_cent.keys())[:10]:
    print(list(dfplayer[dfplayer['Player_Id']==pid]['Player_Name']))
```

```python
momteam = []
mompid = dfmatch['Man_Of_The_Match_Id']
mommid = dfmatch['Match_Id']
momzip = zip(mompid,mommid)
for pid,mid in momzip:
    momteam.extend(list(dfplayer_match[(dfplayer_match['Player_Id']==pid) & (dfplayer_match['Match_Id']==mid)]['Team_Id']))
 
```

```python
momteam
```

```python
momcount=dict([(tid,momteam.count(tid)) for tid in momteam])
momcount
```

```python
mom_keys = list(momcount.keys())
sorted_vals = list(momcount.values())
sorted_vals.sort(reverse=True)
sorted_mom = {}
for val in sorted_vals:
    for key in mom_keys:
        if momcount[key]==val:
            sorted_mom[key]=val
```

```python
print('top 10 teams Based on winning the man of the match')
for tid in sorted_mom.keys():
    print(list(dfteam[dfteam['Team_Id']==tid]['Team_Name']))
```
